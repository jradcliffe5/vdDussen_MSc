```python
import math

def velocity_difference_two_body(G, m1, m2, r):
    # Calculate r1 and r2
    r1 = r * m2 / (m1 + m2)
    r2 = r * m1 / (m1 + m2)
    
    # Calculate velocities v1 and v2
    v1 = math.sqrt(G * m2 / r)
    v2 = math.sqrt(G * m1 / r)
    
    # Calculate absolute velocity difference
    delta_v = abs(v1 - v2)
    
    return delta_v

# Constants (example values)
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
Msun = 2e30 # kg 
m1 = 1e9 * Msun     # Mass of the first body (e.g., Earth) in kg
m2 = 2e9 * Msun     # Mass of the second body (e.g., Moon) in kg
r_arange = np.arange(1,1000) * 3.086e16       # Distance between the two masses in meters (range set from 1 pc to 1kpc)
r = r_arange[4]
# Calculate velocity difference
delta_v = velocity_difference_two_body(G, m1, m2, r) / 1000. 
print(f"Velocity difference: {delta_v:.2f} km/s")
```

```python
delta_v = np.zeros(len(r_arange))
for i,r in enumerate(r_arange):
    delta_v[i] = velocity_difference_two_body(G, m1, m2, r) / 1000.

plt.plot(r_arange/3.086e16,delta_v)
plt.ylabel('radius / pc')
plt.xlabel('delta V / km/s')
```

### model definition

```python
def smbh2_model(p,m1,m2):
    sin3=2/3 #estimate
    numerator = (m1 + m2) * 2 * math.pi * G * sin3
    denominator = p
    cube_root = (numerator / denominator) ** (1/3)
    delta_v = cube_root - 2 * v1

    return delta_v
```

```python
m1_msun = 1e9 # solar masses
m1_kg = m1_msun *  1.989e+30
m2_kg = m1_kg
p_years = 10**np.arange(3,6,0.1) # 1000 to 1 million years
p_sec = p_years * 365.25 * 24 * 60 * 60
```

```python
delta_V_range = smbh2_model(p_sec,m1_kg,m2_kg)
```

```python
plt.plot(delta_V_range,p_years/1e6)
plt.ylabel('period / Myr')
plt.xlabel('delta V / km/s')
```

```python






# Define model (replace with actual mass & orbital period estimation model)
#def smbh_model(theta, m2, p, sini3= 2/3): (this function finds an estimate for vr, which is not what we need. We have vr)
   # m2, p = theta
   # vr= p/(m2*2.0*math.pi*G)*(((v1+v2)**3.0)/(2.0/3.0))-1.0
   # return model

#p=(vr+1)*(2*np.pi*G*m2)*2/(3*(v1+v2)**3)


#This def takes vr (known) and randomly generated values of p and m2 (within a range) and outputs t
#def smbh_model(p,m2):
    
   # v_model= (p/(m2*2.0*np.pi*G))*(((v1+v2)**3.0)/(2.0/3.0))-1.0
    
                                 #mathematical equation derived in olive book
   # return  v_model


# Define prior distributions (replace with your knowledge about mass & period)
#add v prior so model can compare its values to it.
vr_prior = norm( loc=1144.188354, scale=10)
mass1_prior = norm(loc=1e41, scale=1e40)  # Assuming a normal prior for mass with mean 1e41 kg (approx 5e10 M) standard dev: random
mass2_prior = norm(loc=1e42, scale=1e40)
period_prior = norm(loc=8e7, scale=1e6)  # Assuming a normal prior for period with mean 1e5 years and std 1e4 years

# Monte Carlo loop
n_simulations = 10000  # Number of simulations
masses = np.zeros(n_simulations)
periods = np.zeros(n_simulations)
accepted = 0

for i in range(n_simulations):
    # Draw samples from priors
    m2 = mass2_prior.rvs()
    m1 = mass1_prior.rvs()
    p = period_prior.rvs()
    vrp=vr_prior.rvs()

    # Simulate OIII line data with the model
    simulated_v = smbh2_model(p, m1, m2)

    # Likelihood function 
    # This should compare the simulated data with our actual OIII line data (flux vs wavelength)
    # and calculate the probability of getting the observed data given the model parameters
    likelihood = np.exp(-np.sum((vrp - simulated_v)**2))

    # Acceptance probability (Metropolis-Hastings)
    acceptance_ratio = likelihood * period_prior.pdf(p) * mass2_prior.pdf(m2)*mass1_prior.pdf(m1)

    # Random number for acceptance decision
    u = np.random.rand()

    if u < acceptance_ratio:
        masses[i] = m2
        periods[i] = p
        accepted += 1

# Analyze results
accepted_fraction = accepted / n_simulations
print(f"Acceptance fraction: {accepted_fraction:.4f}")

# Extract posterior distribution for mass and period
mass_posterior = np.histogram(masses, bins=20)[0] / n_simulations
period_posterior = np.histogram(periods, bins=20)[0] / n_simulations

# Plot results (optional)
plt.hist(masses, bins=20, label="Mass Posterior")
plt.legend()
plt.show()

plt.hist(periods, bins=20, label="Period Posterior")
plt.legend()
plt.show()

# Find maximum a posteriori (MAP) estimates
map_mass = masses[np.argmax(mass_posterior)]
map_period = periods[np.argmax(period_posterior)]

print(likelihood)
print(acceptance_ratio)
print(vrp)
print(simulated_v)
print(f"MAP Mass: {map_mass:.2e} solar masses")
print(f"MAP Period: {map_period:.2e} years")
```
